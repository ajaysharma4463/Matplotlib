# pie chart


```python
# pie charts shows the share of data like total expences 
```

```python
exp_values=[1400,400,300,20,200]
exp_labels=["house rent","car emi","internt connection","chocklates","other expenses"]
```

```python
plt.pie(exp_values,labels=exp_labels)
```

```python
plt.axis("equal")  # now its look perfect circle
plt.pie(exp_values,labels=exp_labels)
```

```python
plt.axis("equal")  
plt.pie(exp_values,labels=exp_labels)
plt.show() # to get rid of the other information like numeric values
```

```python
# now i want to increase the radius of the circle , by default it is 1
plt.axis("equal")  
plt.pie(exp_values,labels=exp_labels,radius=2)
plt.show()
```

```python
# now i eant o show how much share it si using
plt.axis("equal")  
plt.pie(exp_values,labels=exp_labels,radius=1,autopct="%f%%")  #folating value
plt.show()
```

```python
# but now i want to restrict the value of float 
plt.axis("equal")  
plt.pie(exp_values,labels=exp_labels,radius=1,autopct="%0.2f%%")
plt.show()
# see two decimal points
```

```python
# to see shadow
plt.axis("equal")  
plt.pie(exp_values,labels=exp_labels,radius=1,autopct="%0.2f%%",shadow=True)
plt.show()
```

```python
# now i  want to take off the pies
plt.axis("equal")  
plt.pie(exp_values,labels=exp_labels,radius=1,autopct="%0.2f%%",shadow=True,explode=[0,0.4,0,0.3,0.8]) # it start on th 
#basis of vatiale alligmnet
plt.show()
```

```python
# we can chage the angle
plt.axis("equal")  
plt.pie(exp_values,labels=exp_labels,radius=1,autopct="%0.2f%%",shadow=True,explode=[0,0.4,0,0.3,0.8],startangle=180) 
plt.show()
```

# how to save matplotlib chart in files


```python
plt.axis("equal")  
plt.pie(exp_values,labels=exp_labels,radius=1,autopct="%0.2f%%",shadow=True,explode=[0,0.4,0,0.3,0.8],startangle=180) 
plt.show()
plt.savefig("pngpng.png")
```

```python
# but it is cutting the charts for that
plt.savefig("pngpng.png",bbox_inches="tight")
```

```python
# for extra space or padding
plt.savefig("pngpng.png",bbox_inches="tight",pad_inches=2)
```

```python
# For transpart chart
plt.savefig("pngpng.png",bbox_inches="tight",pad_inches=2, transparent=True)
```
